```python
import spotipy
import spotipy.util as util
import time
import pandas as pd 
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# load spotify.csv

df_full = pd.read_csv('spotify.csv')
df_full = df_full.drop('Unnamed: 0',axis=1)
#df_full.head()
```

# Build playlist dataframe

Response variable (by playlist name):
* followers

Predictors (means):
* acousticness
* danceability 
* duration_ms
* energy
* instrumentalness
* liveness
* loudness
* popularity
* speechiness
* tempo
* valence

Predictors (votes):
* collaborative 
* explicit
* mode
* time_signature

Other predictors?
* artist
* genre
* key

```python
# choose predictors

cont_predictors = ['acousticness','danceability','duration_ms','energy','instrumentalness','liveness','loudness','popularity','speechiness','tempo','valence']
cat_predictors = ['collaborative','explicit','mode','time_signature']
other_predictors = ['artist','genre','key']
response = ['followers']
index = ['playlist_name']
```

```python
# group continuous predictors by playlist -- mean

df_cont = pd.DataFrame()
for pred in cont_predictors:
    df_cont[pred] = df_full[pred]
df_cont['playlist_name'] = df_full['playlist_name']
df_cont = df_cont.groupby('playlist_name').agg(lambda x:np.mean(x))
#df_cont.head()
```

```python
# group categorical predictors by playlist -- vote

df_cat = pd.DataFrame()
for pred in cat_predictors:
    df_cat[pred] = df_full[pred]
df_cat['playlist_name'] = df_full['playlist_name']
df_cat['followers'] = df_full['followers']
df_cat = df_cat.groupby('playlist_name').agg(lambda x:x.value_counts().index[0])
#df_cat.head()
```

```python
df = pd.concat([df_cont,df_cat],axis=1)
df.head()
```

# EDA

```python
# popularity

sns.set_context("poster",font_scale=1)
plt.plot(df['popularity'],df['followers'],'.',label='playlists');
plt.xlabel('mean popularity');
plt.ylabel('followers'); plt.yscale('log')
plt.title('followers versus popularity');
plt.legend();
```

```python
# what about mode? 

# mean
mnmode_df = df[['mode','followers']]
mnmode_df = mnmode_df.groupby('mode').agg(lambda x:np.mean(x))

# median
mdmode_df = df[['mode','followers']]
mdmode_df = mdmode_df.groupby('mode').agg(lambda x:np.median(x))

# combine
mode_df = mdmode_df
mode_df['mean_followers'] = mnmode_df['followers']
mode_df['median_followers'] = mdmode_df['followers']
mode_df = mode_df.drop('followers',axis=1)
print(mode_df.head())

index = np.arange(2)
bar_width = 0.4
plt.bar(index,mode_df['mean_followers'].values,label='means')
plt.bar(index,mode_df['median_followers'].values,label='medians')
objects = ['0.0', '1.0']
plt.xticks(index,objects);
plt.xlabel('mode')
plt.ylabel('followers')
plt.title('mean and median followers across playlists, by mode')
plt.legend();
```

Mode seems helpful. If I remember correctly, mode 0 is major. So people tend to prefer songs in major keys?
