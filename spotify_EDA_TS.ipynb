```python
def bars(var,title,xlab=True,color='blue'):
    sort_var = data.groupby(data[var])
    x=sort_var['followers'].mean()
    ticks=np.arange(len(x))
    plt.bar(ticks,x,color=color,alpha=0.5)
    plt.ylabel('Spotify Followers')
    if xlab==True:
        plt.xticks(ticks,list(x.index),rotation='vertical')
        plt.xlabel(var)
    else:
        plt.xticks([]) 
    plt.title(title)
    #plt.show()
```

```python
plt.figure(figsize=(5, 3))
bars('key','Followers by song key')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
bars('time_signature','Followers by time signature')
plt.show()
```

```python
def word_search(var):
    print("Number of playlists mentioning "+str(var)+":",data['playlist_name'].str.contains(var).sum())
    print("Number of track names mentioning "+str(var)+":",data['track_name'].str.contains(var).sum())
    print("Number of descriptions mentioning "+str(var)+":",data['description'].str.contains(var).sum())
    print("Number of ablum titles mentioning "+str(var)+":",data['album'].str.contains(var).sum())
```

```python
# For tracks
word_search('Relax')
print('---')
word_search('relaxing')
print('---')
word_search('Ocean')
print('---')
word_search('Love')
print('---')
word_search('Hits')
print('---')
word_search('Baby')
```

```python
# For albums
word_search('Greatest')
word_search('greatest')
print('--')
word_search('Best')
word_search('best')
print('---')
word_search('Deluxe')
# word_search('deluxe') all zero
print('---')
word_search('feat.')
word_search('Feat.')
print('---')
word_search('Remastered')
# word_search('remastered') almost zero
```

Some of these may be biased towards longer playlists (if some variable is the same for every track on the playlist) - like for playlist title and description

```python
toptop_data = data[np.log(data.followers+1) > np.mean(np.log(data.followers+1)) + 2*np.std(np.log(data.followers+1))]
print(toptop_data.size)
top_data = data[np.log(data.followers+1) > np.mean(np.log(data.followers+1)) + np.std(np.log(data.followers+1))]
print(top_data.size)
bottom_data = data[np.log(data.followers+1) < np.mean(np.log(data.followers+1)) - 2*np.std(np.log(data.followers+1))]
print(bottom_data.size)
```

```python
def common_words(data,N_top=25):
    print('Track Name')
    print(pd.Series(' '.join(data.dropna()['track_name']).lower().split()).value_counts()[:N_top]/len(data))
    print('--' *30)
    print('Album Name')
    print(pd.Series(' '.join(data.dropna()['album']).lower().split()).value_counts()[:N_top]/len(data))
```

```python
common_words(toptop_data,N_top=5)
```

```python
common_words(top_data)
```

```python
common_words(bottom_data)
```

Notice that the prevalence of certain words, like deluxe/feat./remastered is an order of magnitude different between bottom and top followers data.

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.hist(data.followers.values,label='Histogram')
plt.axvline(data.followers.mean(), 0, 1.0, color='red', label='Mean')
plt.xlabel("Followers")
plt.ylabel("Counts")
plt.title("Spotify Followers (public playlists) Histogram")
plt.legend();
plt.subplot(1,2,2)
plt.hist(np.log(data.followers.values+1),label='Log Histogram',color=sns.color_palette()[1])
plt.axvline(np.mean(np.log(data.followers.values+1)), 0, 1.0, color='red', label='Mean')
plt.xlabel("Followers")
plt.ylabel("Counts")
plt.title("Log Spotify Followers (public playlists) Histogram")
plt.legend();
```
